```python
for value in airline_list:
    valuecountA = seats.loc[seats['operating_airline'] == value, 'departure_country_iata'].value_counts()
    valuecountB = seats.loc[seats['operating_airline'] == value, 'arrival_country_iata'].value_counts()
    
    valuecounts = valuecountA + valuecountB
    most_common_value = valuecounts.idxmax()
    # print(f"The most common value for {value} is {most_common_value}")
```

```python
most_common_values = []
for value in airline_list:
    valuecountA = seats.loc[seats['operating_airline'] == value, 'departure_country_iata'].value_counts()
    valuecountB = seats.loc[seats['operating_airline'] == value, 'arrival_country_iata'].value_counts()
    
    valuecounts = valuecountA + valuecountB
    most_common_value = valuecounts.idxmax()
    most_common_values.append(most_common_value)
    # print(most_common_values)
    # most_common_values[value] = most_common_value
    # print(most_common_value)
```

```python
# add the list of most common values as a new column to the DataFrame
df['country'] = most_common_values
# df = df.drop('country', axis = 1)
print(df)
```

```python
# create a dictionary mapping iata codes to iso codes
iata_to_iso = dict(zip(country['iata'], country['iso']))

# create a new column 'iso' in the seats dataframe, with values from the iata_to_iso dictionary
df['iso'] = df['origin_country'].map(iata_to_iso)
print(df)
```

```python
# Add Airline Type

# create a dictionary mapping iata codes to iso codes
type_airline = dict(zip(airlines['airline_code_iata'], airlines['airline_type']))

# create a new column 'iso' in the seats dataframe, with values from the iata_to_iso dictionary
df['type'] = df['airline'].map(type_airline)
print(df)
```

```python
# Add Full Country Name for Covid dataset

# create a dictionary mapping iata codes to iso codes
fullname_covid = dict(zip(covid['iso_code'], covid['location']))

# create a new column 'fullname' in the seats dataframe, with values from the iata_to_iso dictionary
df['country_covid'] = df['iso'].map(fullname_covid)
# df['fullname'] = df['fullname'].str.replace(r' \(.*\)', '')
print(df)
```
