```python
# seats['type'] = 

# create a dictionary mapping iata codes to iso codes
type = dict(zip(airlines['airline_code_iata'], airlines['airline_type']))

# create a new column 'fullname' in the seats dataframe, with values from the iata_to_iso dictionary
seats['type'] = seats['operating_airline'].map(type)
print(seats)
```

```python
seats['date'] = seats.apply(lambda x: pd.to_datetime(str(x['seats_year']) + '-' + str(x['seats_month'])), 
                   axis=1).dt.strftime('%Y-%m')
print(seats)
```

```python
group_seats = seats.groupby('operating_airline').apply(lambda x: x.sort_values('date', ascending=True))
group_seats
```
